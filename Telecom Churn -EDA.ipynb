```python
telco_data = telco_base_data.copy()
```

```python
#Converting Total Charges to numeric data type
telco_data.TotalCharges = pd.to_numeric(telco_data.TotalCharges, errors='coerce')
telco_data.isnull().sum()
```

```python
telco_data[telco_data['TotalCharges'].isnull() == True]
```

```python
#Since the % of the records compared to total dataset is very low i.e. 0.15%,it is safe to remove them.

telco_data.dropna(inplace=True)
```

```python
telco_data.shape
```

```python
# Max value in tenure column
telco_data['tenure'].max()
```

```python
# Converting the values in tenure column into bins of 12 months

labels = ["{0} - {1}".format(i, i+11) for i in range(1, 72, 12)]

telco_data['tenure_group'] = pd.cut(telco_data.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
telco_data['tenure_group'].value_counts()
```

```python
#Removing column customerID and tenure

telco_data.drop(columns=['customerID', 'tenure'], axis=1, inplace=True)
```

# Univariate Analysis

```python
for i, predictor in enumerate(telco_data.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=telco_data, x=predictor, hue='Churn')
```

```python
#Convert the target variable "Churn" in a binary numeric variable

telco_data['Churn'] = np.where(telco_data.Churn == "Yes", 1, 0)
```

```python
telco_data.head()
```

```python
#Convert all the categorical variables into dummy variables

telco_data_dummies = pd.get_dummies(telco_data, drop_first=True)
telco_data_dummies.head()
```

```python
#Relationship between Monthly Charges and Total Charges

sns.lmplot(data=telco_data_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

```python
# Churn by Monthly Charges and Total Charges

Mth = sns.kdeplot(telco_data_dummies.MonthlyCharges[(telco_data_dummies["Churn"] == 0)],
                 color ="Red", shade = True)
Mth = sns.kdeplot(telco_data_dummies.MonthlyCharges[(telco_data_dummies["Churn"] == 1)],
                 ax=Mth, color='Blue', shade = True)
Mth.legend(['No Churn', 'Churn'], loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly Charges by churn')
```

Insight: Churn is high when Monthly Charges are high

```python
# Build a correlation of all predictors with Churn
plt.figure(figsize=(20,8))
telco_data_dummies.corr()['Churn'].sort_values(ascending=False).plot(kind='bar')
```

# Bivariate Analysis

```python
new_df_target0 = telco_data.loc[telco_data['Churn'] == 0]
new_df_target1 = telco_data.loc[telco_data['Churn'] == 1]
```

```python
def uniplot(df, col, title, hue=None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams['axes.labelsize'] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4 * len(temp.unique())
    fig.set_size_inches(width, 8)
    plt.xticks(rotation = 45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x = col, order = df[col].value_counts().index, hue = hue, palette = 'bright')
    
    plt.show()
```

```python
uniplot(new_df_target1, col='Partner', title="Distrinution of Gender for Churned Customers", hue='gender')
```

```python
uniplot(new_df_target0, col = 'Partner', title = "Distribution of Gender for Non Churned Customers", hue ='gender')
```

```python
uniplot(new_df_target1, col= 'PaymentMethod', title='Distribution of Payment Method for Churned Customers', hue='gender')
```

```python
uniplot(new_df_target1, col='Contract', title="Distribution of Contract for Churned Customers", hue ='gender')
```

```python
uniplot(new_df_target1, col='SeniorCitizen', title="Distribution of Senior Citizen for Churned Customers", hue ='gender')
```
